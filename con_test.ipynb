```python
class ConcertTicket:
    def __init__(self, concert_name, concert_date, ticket_price, available_seats):
        self.concert_name = concert_name
        self.concert_date = concert_date
        self.ticket_price = ticket_price
        self.available_seats = available_seats

class ConcertBookingSystem:
    def __init__(self):
        self.concerts = []

    def add_concert(self, concert):
        self.concerts.append(concert)

    def display_concerts(self):
        print("Available Concerts:")
        for idx, concert in enumerate(self.concerts):
            print(f"{idx + 1}. {concert.concert_name} ({concert.concert_date}) - Price: {concert.ticket_price} Baht - Available Seats: {concert.available_seats}")

    def book_ticket(self, concert_idx, num_tickets):
        concert = self.concerts[concert_idx - 1]
        if concert.available_seats >= num_tickets:
            concert.available_seats -= num_tickets
            total_price = concert.ticket_price * num_tickets
            print(f"Successfully booked {num_tickets} tickets for {concert.concert_name} on {concert.concert_date}. Total Price: {total_price} Baht")
        else:
            print("Sorry, there are not enough available seats.")

# เรียกใช้งาน
if __name__ == "__main__":
    booking_system = ConcertBookingSystem()


    concert1 = ConcertTicket("Concert A", "2024-04-10", 500, 100)
    concert2 = ConcertTicket("Concert B", "2024-05-15", 600, 80)
    concert3 = ConcertTicket("Concert C", "2024-06-20", 700, 120)

    booking_system.add_concert(concert1)
    booking_system.add_concert(concert2)
    booking_system.add_concert(concert3)

    # แสดงรายการคอนเสิร์ต
    booking_system.display_concerts()

    # จองตั๋ว
    concert_choice = int(input("Enter the concert number you want to book: "))
    num_tickets = int(input("Enter the number of tickets you want to book: "))

    booking_system.book_ticket(concert_choice, num_tickets)
```
